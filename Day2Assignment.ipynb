```
#list
list1=[1,2,3,4]
print(type(list1))

#Append
list1.append (5)
print(list1)

#Insert
list1.insert(2,"Hy")
print("\n",list1)

#Extend
list2=[5,5,6,6,6,5,7,8,9]
list1.extend(list2)
print("\n",list1)
#Count
print("\nNumber of 5 in the List are:",list1.count(5))

#len
print("\nLength of the List is:",len(list1)) 

#Dictonary
dicx={"Fname":"Amisha","Lname":"Sharma","Age":"18",1:1,2:2}
print(type(dicx))
#get
print("Accessing a element using get:") 
print(dicx.get("Age"))

#pop
pop_ele1 = dicx.pop(2) 
print("\nDictionary after deletion: "+ str(dicx))
print("Deleted Pair: " + str(pop_ele1))

#popitem
pop_ele2 = dicx.popitem() 
print("\nDictionary after deletion: " + str(dicx)) 
print("Deleted Pair: " + str(pop_ele2))

#clear
dicx.clear() 
print("\nDeleting Entire Dictionary: ") 
print(dicx)

#sets
set1 = {1, 2, 4, 4, 3, 3, 3, 6, 5}
print(type(set1)) 

#ADD
set1.add(6) 
set1.add(7) 
set1.add((8,9)) 
print("\nSet after Addition of elements: ") 
print(set1) 

#update
set1.update([10, 11]) 
print("\nSet after Update: ") 
print(set1) 

#Remove
set1.remove(3) 
set1.remove(4) 
print("\nSet after Removal of elements: ") 
print(set1)

#Discard
set1.discard(10) 
set1.discard(2) 
print("\nSet after Discarding elements: ") 
print(set1) 

#pop
num=set1.pop() 
print("\nSet after popping an element: ") 
print(set1) 
print("Popped number is:",num)

#Tuples
Tuple1 = (18,18,12,3,18,9,1,9,18,20,-1) 
print(type(Tuple1))

#count
print("\nTotal 18 in the Tuple:",Tuple1.count(18))

#index
print("\nIndex of 12:",Tuple1.index(12))

#length
print("\nLength of the Tuple is:",len(Tuple1)	)

#max
print("\nMax element in the Tuple is:",max(Tuple1))

#min
print("\nMin element in the Tuple is:",min(Tuple1))

#Strings
String1 = "hello my name is amisha"
print(type(String1))

#length
print("\nLength of the String is:",len(String1))

#max
print("\nMax element in the Tuple is:",max(String1))

#min
Str="amisha"
print("\nMin element in the Tuple is:",min(Str))

#capatilize
str1="my"
str2="name"
str3="is"
str4="amisha"
print("Capitalize:",str1.capitalize()," ",str2.capitalize()," ",str3.capitalize()," ",str4.capitalize())

#Lower
str="AMISHA"
print("Lower:",str.lower())

```
